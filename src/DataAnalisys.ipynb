```python
import json 
with open('../data/The_ACRE_Crop-Weed_Dataset/data/ACRE_COCO_annotations.json') as json_file:
    data = json.load(json_file)
```

```python
data_info = {}
for ann in data["annotations"]:
    category_id = ann["category_id"]
    plant_name = ann["attributes"]["plant_name"]
    if category_id not in data_info:
        data_info[category_id] = {}
    if plant_name in data_info[category_id]:
        data_info[category_id][plant_name] += 1
    else:
        data_info[category_id][plant_name] = 1

# print(ann,"\n\n")

for k,v in data_info.items():
    print(f"Category id: {k}")
    for k1, v1 in data_info[k].items():
        print("  ", k1, v1)
    print()
```
